```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
correlation =pa.corr()
correlation
```

```python
def heat_map():
    plt.figure(figsize=(8,8),dpi=100)
    sns.heatmap(correlation, annot = True)
    plt.title('heat map for every correlation ' ,fontsize= 20, pad =40)
    plt.savefig('heatmap.png')
```

```python
heat_map()
```

```python
data= pa.head(n=100)
data
```

```python
def bar():
    x= data['County'].value_counts()
    x.plot(kind='bar',figsize=(10,10))
    plt.xlabel('county',fontsize=20)
    plt.ylabel('adoption of electric vehicles ',fontsize=20)
    plt.title('bar plot represent count of individual country',fontsize=20)
    plt.savefig('bar.png')
```

```python
bar()
```

```python
yeardata=data.groupby('Model Year')
for year,year_df in yeardata:
            print(year)
            print(year_df)
```

```python
yeardata01= yeardata.mean('Model Year')
yeardata01
```

```python
finalout=yeardata01.reset_index()
finalout
```

```python
def scatter():
    x = finalout['Model Year']
    y= finalout['Electric Range']
    plt.figure(figsize=(10,10),dpi=100)
    plt.scatter(x,y)
    plt.xlabel('model year' ,fontsize=20)
    plt.ylabel('electric range',fontsize=20)
    plt.title('scatter plot for model year and elecric range')
    plt.savefig('scatter.png')
```

```python
scatter()
```
